```python
df_train['Bed Grade'] = df_train['Bed Grade'].fillna(method="ffill",axis=0)
df_train['City_Code_Patient'] = df_train['City_Code_Patient'].fillna(method="ffill",axis=0)
```

```python
df_test['Bed Grade'] = df_test['Bed Grade'].fillna(method="ffill",axis=0)
df_test['City_Code_Patient'] = df_test['City_Code_Patient'].fillna(method="ffill",axis=0)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_train['Hospital_code'] = le.fit_transform(df_train['Hospital_code'])
df_train['Hospital_type_code'] = le.fit_transform(df_train['Hospital_type_code'])
df_train['Citydf_trainspital'] = le.fit_transform(df_train['City_Code_Hospital'])
df_train['Hospital_region_code'] = le.fit_transform(df_train['Hospital_region_code'])
df_train['Available Extra Rooms in Hospital'] = le.fit_transform(df_train['Available Extra Rooms in Hospital'])
df_train['Department'] = le.fit_transform(df_train['Department'])
df_train['Ward_Type'] = le.fit_transform(df_train['Ward_Type'])
df_train['Ward_Facility_Code'] = le.fit_transform(df_train['Ward_Facility_Code'])
df_train['Bed Grade'] = le.fit_transform(df_train['Bed Grade'])
df_train['City_Code_Patient'] = le.fit_transform(df_train['City_Code_Patient'])
df_train['Type of Admission'] = le.fit_transform(df_train['Type of Admission'])
df_train['Severity of Illness'] = le.fit_transform(df_train['Severity of Illness'])
df_train['Visitors with Patient'] = le.fit_transform(df_train['Visitors with Patient'])
df_train['Age'] = le.fit_transform(df_train['Age'])
```

```python
df_test['Hospital_code'] = le.fit_transform(df_test['Hospital_code'])
df_test['Hospital_type_code'] = le.fit_transform(df_test['Hospital_type_code'])
df_test['Citydf_trainspital'] = le.fit_transform(df_test['City_Code_Hospital'])
df_test['Hospital_region_code'] = le.fit_transform(df_test['Hospital_region_code'])
df_test['Available Extra Rooms in Hospital'] = le.fit_transform(df_test['Available Extra Rooms in Hospital'])
df_test['Department'] = le.fit_transform(df_test['Department'])
df_test['Ward_Type'] = le.fit_transform(df_test['Ward_Type'])
df_test['Ward_Facility_Code'] = le.fit_transform(df_test['Ward_Facility_Code'])
#df_train['Bed Grade'] = le.fit_transform(df_train['Bed Grade'])
df_test['City_Code_Patient'] = le.fit_transform(df_test['City_Code_Patient'])
df_test['Type of Admission'] = le.fit_transform(df_test['Type of Admission'])
df_test['Severity of Illness'] = le.fit_transform(df_test['Severity of Illness'])
df_test['Visitors with Patient'] = le.fit_transform(df_test['Visitors with Patient'])
df_test['Age'] = le.fit_transform(df_test['Age'])
```

```python
from sklearn import preprocessing
mm_scaler = preprocessing.MinMaxScaler()
df_train[['Admission_Deposit']] = mm_scaler.fit_transform(df_train[['Admission_Deposit']])
df_test[['Admission_Deposit']] = mm_scaler.fit_transform(df_test[['Admission_Deposit']])
```

```python
print(df_test.head(3))
```

```python
x_train = df_train
x_train = x_train.drop(['case_id'], axis=1)
x_train = x_train.drop(['Stay'], axis=1)
y_train = df_train['Stay']

x_test = df_test
x_test = x_test.drop(['case_id'], axis=1)
```

```python
import lightgbm as lgb
lgb_cl = lgb.LGBMClassifier()

#from sklearn.svm import SVC
#svm_cl = SVC()
#from sklearn.ensemble import GradientBoostingClassifier
#gb_cl = GradientBoostingClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train_spl, x_test_spl, y_train_spl, y_test_spl = train_test_split(x_train, y_train, test_size=0.20, random_state=8)
```

```python
lgb_cl.fit(x_train_spl, np.ravel(y_train_spl))
#svm_cl.fit(x_train_spl, np.ravel(y_train_spl))
#gb_cl.fit(x_train_spl, np.ravel(y_train_spl))
```

```python
y_test_pred_lgb = lgb_cl.predict(x_test_spl)
#y_test_pred_svm = svm_cl.predict(x_test_spl)
#y_test_pred_gb = gb_cl.predict(x_test_spl)
```

```python
from sklearn.metrics import accuracy_score
#print("svm accuracy: "accuracy_score(y_test_spl, y_test_pred_svm))
print("lgb accuracy: ", accuracy_score(y_test_spl, y_test_pred_lgb))
#print("gb accuracy: "accuracy_score(y_test_spl, y_test_pred_gb))
```

```python
final_pred = lgb_cl.predict(x_test)
```
